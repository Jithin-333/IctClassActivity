```python
FILE_NAME = "expenses.txt"



def add_expense():
    category = input("Enter category: ")

    if category == "":
        print("Category cannot be empty")
        return

    amount_input = input("Enter amount: ")

    try:
        amount = float(amount_input)
    except:
        print("Please enter a valid number")
        return

    if amount <= 0:
        print("Amount must be greater than 0")
        return


    file = open(FILE_NAME, "a")
    file.write(category + "," + str(amount) + "\n")
    file.close()

    print("Expense added successfully!")



def view_expenses():
    try:
        file = open(FILE_NAME, "r")
        lines = file.readlines()
        file.close()

        if len(lines) == 0:
            print("No expenses found")
            return

        print("\nYour Expenses:\n")

        count = 1
        for line in lines:
            data = line.strip().split(",")

            if len(data) == 2:
                category = data[0]
                amount = float(data[1])

                print(str(count) + ". " + category + " - Rs." + str(amount))
                count += 1

    except:
        print("No file found. Add expenses first.")



def calculate_total():
    try:
        file = open(FILE_NAME, "r")
        lines = file.readlines()
        file.close()

        if len(lines) == 0:
            print("No expenses to calculate")
            return

        total = 0

        for line in lines:
            data = line.strip().split(",")

            if len(data) == 2:
                amount = float(data[1])
                total += amount

        print("\nTotal Expense = Rs.", total)

    except:
        print("No file found.")



def expense_tracker():
    while True:
        print("\n====== EXPENSE TRACKER ======")
        print("1. Add Expense")
        print("2. View Expenses")
        print("3. Total Expense")
        print("4. Exit")

        choice = input("Enter choice: ")

        if choice == "1":
            add_expense()

        elif choice == "2":
            view_expenses()

        elif choice == "3":
            calculate_total()

        elif choice == "4":
            print("Goodbye!")
            break

        else:
            print("Invalid choice. Try again.")


expense_tracker()
```
